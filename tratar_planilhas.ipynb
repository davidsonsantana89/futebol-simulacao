## PARTIDAS DISPUTADAS

```python
caminho_realizadas = '/content/drive/MyDrive/PROJETOS PESSOAIS/SIMULADOR BRASILEIRAO/FLASHSCORE/base-dados/brazil/serie-a/serie-a-2024-com-minutos.csv'
```

```python
df_realizadas = pd.read_csv(caminho_realizadas, sep=';')
df_realizadas.tail()
```

```python
id_realizadas = df_realizadas['id_jogo']
# id_realizadas
```

```python
df_realizadas['dia'] = pd.to_datetime(df_realizadas['dia'], format='mixed', dayfirst=True, errors='coerce')
```

```python
df_realizadas = df_realizadas[['dia', 'home', 'away','FTHG', 'FTAG']]
df_realizadas = df_realizadas.copy()
df_realizadas.rename({'dia':'data'}, axis=1, inplace=True)
df_realizadas.head()
```

```python
df_realizadas.to_csv('2024-serie-a-passado_20240615.csv',index=False)
```

## PRÓXIMAS PARTIDAS

```python
caminho_fixtures = '/content/drive/MyDrive/PROJETOS PESSOAIS/SIMULADOR BRASILEIRAO/FLASHSCORE/base-dados/brazil/serie-a/fixtures/serie-a-2024-fixtures.csv'
```

```python
df_fixtures = pd.read_csv(caminho_fixtures, sep=';')
# df_fixtures = df_fixtures[:-2]
df_fixtures = df_fixtures
df_fixtures.tail()
```

```python
df_fixtures = df_fixtures.query('id_jogo not in @id_realizadas')
df_fixtures.tail()
```

```python
df_fixtures = df_fixtures[['dia', 'home', 'away']]
```

```python
df_fixtures = df_fixtures.copy()
df_fixtures['dia'] = pd.to_datetime(df_fixtures['dia'], format='mixed', dayfirst=True)
```

```python
df_fixtures.sort_values(by='dia', ascending=True, inplace=True)
df_fixtures.reset_index(drop=True, inplace=True)
```

```python
df_fixtures.rename({'dia':'data'}, axis=1, inplace=True)
```

```python
df_fixtures.head()
```

```python
df_fixtures.to_csv('2024-serie-a-fixtures_20240615.csv',index=False)
```
